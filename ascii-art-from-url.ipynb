```python
# https://www.hackerearth.com/notes/beautiful-python-a-simple-ascii-art-generator-from-images/

%matplotlib inline

import urllib, cStringIO
import math
from PIL import Image as Image
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

from __future__ import division
from IPython.html.widgets import interact
```

```python
def show_image(image):
    fig, ax = plt.subplots()
    plt.figure(figsize=(3,3))
    plt.gray()

    ax.imshow(image)
    ax.axis('off')

    plt.show()
    image_as_array = np.asarray(image)
    print image_as_array.size
```

```python
def show_images(images):
    num_images = len(images)
    fig, axes = plt.subplots(ncols=num_images)
    plt.figure(figsize=(3,9))
    plt.gray()

    for i in range(num_images):
        axes[i].imshow(images[i])
        axes[i].axis('off')
    
    plt.show()
```

```python
def make_to_ascii(image_url_path, ASCII_CHARS, new_width=35):
    file = cStringIO.StringIO(urllib.urlopen(image_url_path).read())
    image = Image.open(file)

    (original_width, original_height) = image.size
    aspect_ratio = original_height/float(original_width)
    new_height = int(aspect_ratio * new_width)

    small_image = image.resize((new_width, new_height))
    
    gray_image = small_image.convert('L')

    pixels_in_image = list(gray_image.getdata())
    range_width=256/(len(ASCII_CHARS)-1)
    pixels_to_chars = [ASCII_CHARS[int(math.floor(pixel_value/range_width))] for pixel_value in pixels_in_image]
    pixels_to_chars_string = "".join(pixels_to_chars)
    ascii_width = new_width * len(ASCII_CHARS[0])
    image_ascii = [pixels_to_chars_string[index: index + ascii_width] for index in xrange(0, len_pixels_to_chars, ascii_width)]
    ascii_art = "\n".join(image_ascii)
    show_images([image, small_image, gray_image])
    print(ascii_art)
```
